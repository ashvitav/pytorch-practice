```python
class Dog:
   """This is a dog class"""
   age = 5

   def __init__(self, name, breed, color):
       self.name = name
       self.breed = breed
       self.color = color

   def get_color(self):
       return f"Color is {self.color}"   #The f-prefix and curly braces allow embedding expressions inside string literals

   def bark(self):
       return "Woof!"
   
   def get_info(self):
       return f"{self.name} is a {self.breed} and is {self.age} years old"
   


# Create a Dog object
my_dog = Dog("Max", "Golden Retriever","yellow")
print(my_dog.get_color())
print(my_dog.get_info())
print(my_dog.bark())


dog_pairs = [
   ("Luna", "Husky"),
   ("Rocky", "German Shepherd"),
   ("Bella", "Labrador"),
   ("Charlie", "Poodle"),
   ("Daisy", "Corgi"),
   ("Bear", "Bernese Mountain Dog"),
   ("Molly", "Beagle"),
   ("Duke", "Great Dane")
]
```

```python
class SmallDog(Dog):
    def __init__(self, name, breed,color):
       super().__init__(name, breed, color)
       self.toys = "ball"
    
    def make_sound(self):
       return "Woof! Woof! Woof!"

SmallDog2 = SmallDog("Max", "Golden Retriever","yellow")
print(SmallDog2.toys)
SmallDog2.make_sound()
```

```python
class GuardDog(Dog):
   """Guard dogs have special security methods"""
   def bark(self):
       return "WOOF! WOOF!"
   
   def guard(self):
       return f"{self.name} is guarding the house"

# Create instances
yorkie = SmallDog("Pixie", "Yorkshire Terrier","black")
shepherd = GuardDog("Rex", "German Shepherd","white")
```

```python
class ComplexNumber:
  def __init__(self, r=0, i=0):
    self.real = r
    self.imag = i

  def get_data(self):
    print(f'{self.real}+{self.imag}j')

# Create a new object
num1 = ComplexNumber(2, 3) 
# Call method
num1.get_data()
```

```python
import torch
import torch.nn as nn

class SimpleNN(nn.Module):
   def __init__(self):
       super(SimpleNN, self).__init__()
       self.linear1 = nn.Linear(2, 4)  # 2 inputs -> 4 hidden
       self.relu = nn.ReLU()
       self.linear2 = nn.Linear(4, 1)  # 4 hidden -> 1 output

   def forward(self, x):
       x = self.linear1(x)
       x = self.relu(x)
       x = self.linear2(x)
       return x

# Create model and test
model = SimpleNN()
input_data = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
output = model(input_data)
print(output)
```

## Python basics

```python
# Numbers: Integers and Floats
age = 25                  # Integer (whole number)
height = 1.75            # Float (decimal number)
complex_num = 3 + 4j     # Complex number

# Strings (text)
name = "Alice"           # Can use single or double quotes
message = 'Hello there'  
multi_line = """This is a
multi-line string"""     # Triple quotes for multiple lines

# Boolean (True/False)
is_student = True
has_license = False

# Lists (ordered, mutable collection)
fruits = ["apple", "banana", "orange"]
mixed_list = [1, "hello", True, 3.14]
fruits[0] = "pear"      # Lists can be modified

# Tuples (ordered, immutable collection)
coordinates = (10, 20)   # Can't be modified after creation
rgb_color = (255, 128, 0)

# Dictionaries (key-value pairs)
person = {
    "name": "Bob",
    "age": 30,
    "city": "New York"
}
person["age"] = 31      # Values can be modified

# Sets (unordered collection of unique elements)
numbers = {1, 2, 3, 4, 4}  # Note: duplicate 4 will be removed
colors = {"red", "blue", "green"}

# None (represents absence of value)
empty_variable = None

# Example of checking variable types
print(type(age))        # <class 'int'>
print(type(height))     # <class 'float'>
print(type(name))       # <class 'str'>
print(type(fruits))     # <class 'list'>
print(type(person))     # <class 'dict'>


# Function definition examples (Python uses indentation to define code blocks.s)
def greet(name):
   return f"Hello, {name}!"

def calculate_total(price, tax_rate=0.1):
   return price + (price * tax_rate)

# Function with multiple return values
def get_stats(numbers):
   return min(numbers), max(numbers), sum(numbers)

# If statement examples
age = 18
if age >= 18:
   print("Adult")
elif age >= 13:
   print("Teenager") 
else:
   print("Child")

# For loop examples
# Loop through list
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
   print(fruit)

# Range loop
for i in range(5):
   print(i)  # Prints 0 to 4

# Loop with enumerate
for index, fruit in enumerate(fruits):
   print(f"{index}: {fruit}")

# Nested loops with conditional
matrix = [[1,2,3], [4,5,6]]
for row in matrix:
   for num in row:
       if num % 2 == 0:
           print(f"{num} is even")
       else:
           print(f"{num} is odd")
```
